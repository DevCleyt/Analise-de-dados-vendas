```python
# 1º importar base de dados
import win32com.client as win32
import pandas as pd
import pandas as pd
tabela_vendas = pd.read_excel('Vendas.xlsx')
pd.set_option('display.max_columns', None)
```

```python
# 2º Visualizar a base de dados

tabela_vendas[['ID Loja','Valor Final']]
```

```python
# Faturamento por Loja

faturamento = tabela_vendas[['ID Loja','Valor Final']].groupby('ID Loja').sum()
```

```python
print(faturamento)
```

```python
# 4º Quantidade de produto vendido por loja
tabela_vendas.head()
```

```python
quantidade = tabela_vendas[['ID Loja','Quantidade']].groupby('ID Loja').sum()
```

```python
print(quantidade)
```

```python
# 5º Ticket medio por produto em cada loja ( Calcular Faturamento / por quantidade de produt vendido  )

ticket_medio = (faturamento ['Valor Final'] / quantidade['Quantidade']).to_frame()
ticket_medio = ticket_medio.rename(columns={0:'ticket_medio'})
```

```python
print(ticket_medio)
```

```python
# 6º Enviar e-mail com relatorio.

outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)
mail.To = 'pereiradossantoscleyton@gmail.com'
mail.Subject = 'Analise de dados, Relatório de Vendas por Loja shopping'
mail.HTMLBody = f'''

<p>Prezado Cleyton, boa tarde !!</p>
<p>Segue relatorio de vendas por cada loja. </p>

<p>Faturamento: </p>
{faturamento.to_html()}

<p>Quantidade Vendida: </p>
{quantidade.to_html()}

<p>Ticket-medio dos produtos em cada loja: </p>
{ticket_medio.to_html()}

<p> Att: Cleyton </p>
'''

mail.Send()
```

```python
# Gera o código HTML (exemplo com pandas DataFrame) para abrir na web o resultado do relatorio.

html_code = f"""
<html>
<head><title>Relatório de Vendas</title></head>

<body>
    <h1>Relatório de Vendas</h1>

    <p><strong>Faturamento:</strong></p>
    {faturamento.to_html(formatters={'Valor Final':'R${:,.2F}'.format})}

    <p><strong>Quantidade Vendida:</strong></p>
    {quantidade.to_html(formatters={'quantidade':'{:,.2f}'.format})}

    <p><strong>Ticket Médio:</strong></p>
    {ticket_medio.to_html(formatters={'ticket_medio':'R${:,.2F}'.format})}
</body>
</html>
"""

# Salva o HTML em um arquivo
with open("relatorio_vendas.html", "w", encoding="utf-8") as file:
    file.write(html_code)

# Abre no navegador automaticamente
import webbrowser
webbrowser.open("relatorio_vendas.html")
```
